## PCA

```r
result_s_asum <- pca(asum[, -c(20, 21)])
result_s_asum
```

```r
default <- par()
mai <- par()$mai
mai[4] <- mai[1]
mai = c(1.2, 1.2, 0.68, 1.25)
par(cex=1.5)
par(mai = mai)
plot(result_s_asum$eval, type='o', xlab='PC', ylab="固有値", col='red', lwd=5, ylim=c(0,10))
par(new = T)
plot(numeric(19)+1, type='l',lty="dashed", ylim=c(0,10), lwd=3, xlab='', ylab='')
par(new = T)
plot(result_s_asum$cum.contribution, type='o', pch=4, xlab='PC', axes = FALSE,  ylab="", col='blue', lwd=5, ylim=c(35,100))
par(new = T)
plot(numeric(19)+80, type='l',lty="dashed", ylim=c(35,100), lwd=3, axes = FALSE, xlab='', ylab='')
axis(4)
mtext("累積寄与率 [%]", side=4, line=3, cex=1.5)
par(new = T)
par(mai = default$mai)
dev.copy(bmp, file="scree.bmp")
dev.off()
```

```r
cor(result_s_asum$fs)
```

```r
plot(result_s_asum$fs[,c(1)], result_s_asum$fs[,c(2)], xlab="PC1(好みでない曲)", ylab="PC2(見た目の洗練さ)", cex.lab=1.4, cex.axis=1.2)
text(result_s_asum$fs[, c(1)], result_s_asum$fs[,c(2)], cex=2)
dev.copy(bmp, file="2d-scatter_p1xp2.bmp")
dev.off()
plot(result_s_asum$fs[,c(1)], result_s_asum$fs[,c(3)], xlab="PC1(好みでない曲)", ylab="PC3(他を犠牲にした迫力)", cex.lab=1.4, cex.axis=1.2)
text(result_s_asum$fs[, c(1)], result_s_asum$fs[,c(3)], cex=2)
dev.copy(bmp, file="2d-scatter_p1xp3.bmp")
dev.off()
plot(result_s_asum$fs[,c(2)], result_s_asum$fs[,c(3)], xlab="PC2(見た目の洗練さ)", ylab="PC2(他を犠牲にした迫力)", cex.lab=1.4, cex.axis=1.2)
text(result_s_asum$fs[, c(2)], result_s_asum$fs[,c(3)], cex=2)
dev.copy(bmp, file="2d-scatter_p2xp3.bmp")
dev.off()
```

# pv06

```r
setwd('/mnt/d/workspace/UEC/term_6/media2019/2nd/pv06')
datasum <- data.matrix(read.csv('pv06.csv'))
```

## PCA

```r
result_s <- pca(datasum[,-c(20,21)])
```

```r
result_s
```

```r
default <- par()
mai <- par()$mai
mai[4] <- mai[1]
mai = c(1.2, 1.2, 0.68, 1.25)
par(cex=1.5)
par(mai = mai)
plot(result_s$eval, type='o', xlab='PC', ylab="固有値", col='red', lwd=5, ylim=c(0,7))
par(new = T)
plot(numeric(19)+1, type='l',lty="dashed", ylim=c(0,7), lwd=3, xlab='', ylab='')
par(new = T)
plot(result_s$cum.contribution, type='o', xlab='PC',  pch=4, axes = FALSE,  ylab="", col='blue', lwd=5, ylim=c(35,100))
par(new = T)
plot(numeric(19)+80, type='l', lty="dashed", ylim=c(35,100), lwd=3, axes = FALSE, xlab='', ylab='')
axis(4)
mtext("累積寄与率 [%]", side=4, line=3, cex=1.5)
par(new = T)
par(mai = default$mai)
dev.copy(bmp, file="scree.bmp")
dev.off()
```

CSV書き出し

```r
write.table(result_s$mean, "mean.csv", quote=F, row.names=T, col.names=T, append=F)
write.table(result_s$variance, "variance.csv", quote=F, row.names=T, col.names=T, append=F)
write.table(result_s$standard.deviation, "standard.deviation.csv", quote=F, row.names=T, col.names=T, append=F)
write.table(result_s$r, "r.csv", quote=F, row.names=T, col.names=T, append=F)
write.table(result_s$eval, "eval.csv", quote=F, row.names=T, col.names=T, append=F)
write.table(result_s$contribution, "contribution.csv", quote=F, row.names=T, col.names=T, append=F)
write.table(result_s$cum.contribution, "cum.contribution.csv", quote=F, row.names=T, col.names=T, append=F)
write.table(result_s$factor.loadings, "factor.loadings.csv", quote=F, row.names=T, col.names=T, append=F)
write.table(result_s$fs, "fs.csv", quote=F, row.names=T, col.names=T, append=F)
```
